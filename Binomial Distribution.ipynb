# **Binomial Distribution**

The binomial distribution represents the probability distribution of a random variable that counts the number of successes in a fixed number of independent trials. Each trial results in either success (with probability p ) or failure (with probability 1-p)

Examples:

- Coin Filp
- New subscriber from watching a YOutubeVideo (Subscribe / Not Subscribe)
- Pitcher Throwing a Strike (Strike Ball)

Terms to know

- Success (p) (also known as probability at one triall)
    - 2.5% of views turn into subscribers
        - P(x=1) = 0.025
        - P(x=0) = 0.975

- Trials (n)
    - Video Gets 1,000 views

- Size (Number of Trials)
    - Look at 100 Different videos

## Bernoulli Distribution - Concepts

Binomial Outcomes of multiple trials of an event, bernoulli one trial of an event

PMF (Probability Mass Function) - Exact Probability of a number of events happening

CDF (Cumulative Distribution Function)- Probability Value less or equal to Number of Events

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

from scipy.stats import binom

np.random.seed(12)
```

### Example #2 Coin flip Scipy

```python
p_success_coin = 0.5
n_trials = 50
size_coin_flips = 25

scipy_coins = binom.rvs(n=n_trials, p=p_success_coin, size=size_coin_flips)
scipy_coins
```

### Example #5 Plot Coin Histogram

```python
plt.figure(figsize=(8,6))
plt.hist(scipy_coins, bins=10, color='skyblue', edgecolor="black")
plt.title('Histogram coin flips')
plt.xlabel("Success")
plt.ylabel('Frequency')
plt.show()
```

### Example #5 Plot Coin Flip PMF

```python
x = np.arange(0, n_trials + 1)
pmf = binom.pmf(x, n=n_trials, p=p_success_coin)

plt.figure(figsize=(10,8))
plt.stem(x, pmf)
plt.title("PMF coin flips")
plt.xlabel("Success")
plt.ylabel('Probability')
plt.show()
```

### Example #6 CDF Plot 

```python
x = np.arange(0, n_trials + 1)
cdf = binom.cdf(x, n=n_trials, p=p_success_coin)

plt.figure(figsize=(10,8))
plt.plot(x, cdf,'b-', lw=2)
plt.title("CDF coin flips")
plt.xlabel("Success")
plt.ylabel('Probability')
plt.show()
```
