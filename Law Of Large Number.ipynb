# **Law of Large Number** 

is a fundamental theorem in probability theory and statistics. It describes the result of performing the same experiment a large number of times. The theorem states that as the number of trials or observations increases, the sample mean will get closer and closer to the expected value (population mean). This means that over time, the average of the results obtained from a large number of trials should converge to the expected value.

#### 1. Perform experiemt many times

#### 2. The experiments averages will reach the true average as incrase number of trials

#### 3. Observed probability goes to theoretical probability

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(15)
```

### Example #1 Bernoulli

```python
n_trials = 10000

coin_filps = np.random.randint(0,2, size=n_trials)
cumulative_mean = np.cumsum(coin_filps) / (np.arange(1, n_trials + 1))
cumulative_mean
```

```python
plt.figure(figsize=(10,6))
plt.plot(cumulative_mean, label="Cumulative Mean")
plt.axhline(y=0.5, color='r', linestyle='--', label="Expected Value 0.5")
plt.xlabel("Num of trials")
plt.ylabel("Cumulative Mean")
plt.title("Law of Large Number Coin Flips")
plt.legend()
plt.show()
```

### Example #2 Data Science Salaries (Normal Distribution)

```python
n_samples = 10000
mu = 120000
sigma = 30000
salaries = np.random.normal(mu,sigma, n_samples)
cumulative_mean_salaries = np.cumsum(salaries) / (np.arange(1, n_samples + 1))
cumulative_mean_salaries
```

```python
plt.figure(figsize=(10,6))
plt.plot(cumulative_mean_salaries, label="Cumulative Mean Salary")
plt.axhline(y=mu, color='r', linestyle='--', label="Expected Salary $120,000")
plt.xlabel("Num of samples")
plt.ylabel("Cumulative Mean Salary")
plt.title("Law of Large Number Normal Distribution DS Salaries")
plt.legend()
plt.show()
```
