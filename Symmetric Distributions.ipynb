# **Symmetric Distributions (Left & Right)** 

 is a type of probability distribution in which the left and right sides of the distribution are mirror images of each other around the central point (mean or median). In a symmetric distribution, the values on either side of the central value are equally likely to occur, and the shape of the distribution is the same on both sides.

A skwred distribution is one where the data is not symmetrically distributed around the mean. There are two types of skewed distributions

Right and Left Skwwed also knows as Positively and Negatively Skewed 

### Characteristics:

#### 1. Mean = Median = Mode: In a perfectly symmetric distribution, the mean, median, and mode are all located at the center of the distribution.

#### 2. Shape: The distribution has equal spread on both sides of the center. The values are symmetrically distributed around the center.

#### 3. Mirror Image: The distribution's left half is a mirror image of the right half.

## Right Skewed (Positively Skewed)

##### The right tail (larger values) is longer, meaning the mass of the distruvitions on the left. In such distributions, the mean is typically greater than the median.

Sample: Housing Prices


## Left Skewed (Negatively Skewed)

The left tail (smaller values) is longer, meaning the mass of the distribution is concentrated on the right. In such distributions, thmean is typically less than the median

Sample: Retirement Age

### Symmetric (Normal) Distribution

A symmetric distribution is one where the left and right sides of the distribution are mirror images of each other. As a result, the mean, median, mode are all equal and located at the center of the distribution

Sample: IQ Scores

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
from scipy import stats

np.random.seed(13)
```

```python
right_skewed_data = np.random.exponential(scale=2, size=1000)
left_skewed_data = np.random.exponential(scale=2, size=1000) * -1 + 7
normal_data = np.random.normal(loc = 5, scale=1.5, size=1000)
```

## Calculate Mean Median Mode in each

### Right

```python
mean_right = np.mean(right_skewed_data)
median_right = np.median(right_skewed_data)
mode_right = stats.mode(right_skewed_data, keepdims=True)[0][0]
```

### Left

```python
mean_left = np.mean(left_skewed_data)
median_left = np.median(left_skewed_data)
mode_left = stats.mode(left_skewed_data, keepdims=True)[0][0]
```

### Normal

```python
mean_normal = np.mean(normal_data)
median_normal = np.median(normal_data)
mode_normal = stats.mode(normal_data, keepdims=True)[0][0]
```

### Plot there distributions

```python
plt.figure(figsize=(21,6))

plt.subplot(1,3,1)
sns.histplot(right_skewed_data, kde=True, color="skyblue")
plt.axvline(mean_right, color="r", linestyle="--", label=f"mean: {mean_right}")
plt.axvline(median_right, color="g", linestyle="-", label=f"median: {median_right}")
plt.axvline(mode_right, color="b", linestyle="--", label=f"mode: {mode_right}")
plt.title('Right Skewed Distribution')
plt.legend()


plt.subplot(1,3,2)
sns.histplot(left_skewed_data, kde=True, color="skyblue")
plt.axvline(mean_left, color="r", linestyle="--", label=f"mean: {mean_left}")
plt.axvline(median_left, color="g", linestyle="-", label=f"median: {median_left}")
plt.axvline(mode_left, color="b", linestyle="--", label=f"mode: {mode_left}")
plt.title('Left Skewed Distribution')
plt.legend()


plt.subplot(1,3,3)
sns.histplot(normal_data, kde=True, color="skyblue")
plt.axvline(mean_normal, color="r", linestyle="--", label=f"mean: {mean_normal}")
plt.axvline(median_normal, color="g", linestyle="-", label=f"median: {median_normal}")
plt.axvline(mode_normal, color="b", linestyle="--", label=f"mode: {mode_normal}")
plt.title('Normal Skewed Distribution')
plt.legend()


plt.show()
```

### Box plots

```python
plt.subplot(1,3,1)
sns.boxplot(y=right_skewed_data, color='skyblue')
plt.title("Right-Skewed Distributions")

plt.subplot(1,3,2)
sns.boxplot(y=left_skewed_data, color='lightgreen')
plt.title("Left-Skewed Distributions")

plt.subplot(1,3,3)
sns.boxplot(y=normal_data, color='lightcoral')
plt.title("Normal Skewed Distributions")

plt.tight_layout()
plt.subplots_adjust(wspace=0.6, top=0.8)
plt.show()
```
