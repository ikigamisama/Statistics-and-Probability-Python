# **Geometric Distribution**

is a probability distribution that models the number of trials required to achieve the first success in a series of independent and identically distributed Bernoulli trials (where each trial has two possible outcomes: success or failre)

Examples:

- Test X numbers of items until a defective item is found
- NUmber of views needed for a new subscriber from a video
- Coins Flipped Until tails

Discrete

PMF (Probability Mass function) - Exact Probability that the first success occurs on the k-th trial

CDF (Cumulative Distribution function) - Probability that the firtst success occurs within the first k trials

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import geom

np.random.seed(17)
```

### Example #1 1000 Random Samples with Numpy

```python
p = 0.125
samples = np.random.geometric(p, 1000)
samples
```

### Example #2 Calc Mean, Variance, STD

```python
mean = np.mean(samples)
variance = np.var(samples)
std = np.std(samples)
```

### Example #3 Histograms

```python
plt.figure(figsize=(12,6))

plt.hist(samples, bins = np.arange(1,53) - 0.5, edgecolor="black", rwidth=0.8)
plt.title("Histogram of Geo Data Distribution")
plt.xlabel("Packs Open")
plt.ylabel("Frequency")
plt.xticks(np.arange(1,53))
plt.grid(axis='y', linestyle="--", alpha=0.7)
```

### Example #4 PMF

```python
x = np.arange(1, max(samples) + 1)
pmf = geom.pmf(x, p)


plt.hist(samples, bins = np.arange(0.5, max(samples) + 1.5) , edgecolor="black",  density=True, alpha=0.6, color='Green')
plt.title("Histogram of PMF In Geo Data Distribution")
plt.xlabel("Packs Open")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

### Example #5 CDF

```python
cdf = geom.cdf(x, p)
plt.step(x, cdf, where='post', color='blue')
plt.title("CDF In Geo Data Distribution")
plt.xlabel("Number of Trials")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()
```
