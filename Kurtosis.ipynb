# **Kurtosis**

Thinkness/Shape of a Tale compared to the rest of the distribution <br>
Helps show the extreme values in a datasets

- Mesokurtic
- Leptokurtic
- Platykurtic

Normal Distributions have a kurtosis of 3 

Scipy refers to excess kurtosis which is the kurtosis minus 3. This shifts the normal distribution's kurtosis to 0

## Leptokurtic

Distribution with heavy tails or outliers, which means there are more extreme values than the normal distriubtion

The Distributions has heavier tails, so the kurtosis will be grater than 0 

## Platykurtic

This describes a distriubtion with light tails or fewer outliers compared to a normal distriubtion.

This distribution has lighter tails, so the kurtosis will be less than 0 

## Distribution covered 

- **Normal** - A Bell Curve or Gaussian Distribution is perfectly symmentrical, with the mean, median and mode all being equal, creating a balance shapre with identical left and right sides.

- **Uniform** - A uniform distribution is a type of probability distribution in which all outcomes are equally like within a certain range 

- **Laplace** - The Laplace distribution is symmetric and centered at a mean value, with a sharp peak at the center (mean) and heavy tails

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns 

np.random.seed(11)
```

```python
data_normal = np.random.normal(0, 1, 1000)
data_leptrokurtic = np.random.laplace(0,1, 1000)
data_platykurtic = np.random.uniform(-1,1, 1000)
```

### Example #1 Build Plots

```python
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
sns.histplot(data_normal, kde=True, color="blue")
plt.title("Mesokurtic")

plt.subplot(1,3,2)
sns.histplot(data_leptrokurtic, kde=True, color="red")
plt.title("Leptrokurtic")

plt.subplot(1,3,3)
sns.histplot(data_platykurtic, kde=True, color="green")
plt.title("Platykurtic")
```

### Example #2 QQ Plot

```python
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
stats.probplot(data_normal, dist="norm", plot=plt)
plt.title("Mesokurtic")

plt.subplot(1,3,2)
stats.probplot(data_leptrokurtic, dist="norm", plot=plt)
plt.title("Leptrokurtic")

plt.subplot(1,3,3)
stats.probplot(data_platykurtic,  dist="norm", plot=plt)
plt.title("Platykurtic")
```

### Example #3 Kurtosis Values

```python
kurt_nomal = stats.kurtosis(data_normal)
kurt_leptokurtic = stats.kurtosis(data_leptrokurtic)
kurt_platykurtic = stats.kurtosis(data_platykurtic)
```
