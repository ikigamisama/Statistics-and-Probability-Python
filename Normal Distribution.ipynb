# **Normal Distribution**

Bell Curve/Gaussian Distribution

Emperical Rule 

- 68%   - 1 standard deviation of the mean
- 95%   - 2 standard deviation of the mean
- 99.7% - 3 standard deviation of the mean

The mean, median and mode are exactly the same
Symmetrical both Left and Right

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import norm
```

### Example #1 Generating a Normal Distribution

```python
mean = 0
std_dev = 1
size = 1000

data = np.random.normal(loc=mean, scale=std_dev, size=size)

print(np.mean(data))
print(np.median(data))
print(np.std(data))
```

### Example #2 Plot a histogram

```python
plt.figure(figsize=(10,8))
sns.histplot(data)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### Example #3 Plot a histogram with KDE 

```python
plt.figure(figsize=(10,8))
sns.histplot(data, kde=True)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram With KDE")
plt.show()
```

### Example #4 CDF Probabilities

```python
cdf_neg2 = norm.cdf(-2, loc=data.mean(), scale=data.std())
cdf_neg2
```

```python
cdf_zero = norm.cdf(0, loc=data.mean(), scale=data.std())
cdf_zero
```

```python
cdf_2 = norm.cdf(2, loc=data.mean(), scale=data.std())
cdf_2
```

### Example #5 Range using CDF

```python
lower_bound = -1
upper_bound = 1

cdf_lower = norm.cdf(lower_bound, loc=data.mean(), scale=data.std())
cdf_upper = norm.cdf(upper_bound, loc=data.mean(), scale=data.std())

cdf_range = cdf_upper - cdf_lower

cdf_range
```

### Example #6 Quantiles

```python
Q1 = np.percentile(data, 25)
Q1
```
