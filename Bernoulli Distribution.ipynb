# **Bernoulli Distribution**

The Bernouilli distribution represents the probability of a random variable that takes on one of two outcomes: success (with probability p) or failure (with probability 1-p) in a single trial. Unline the binomial distribution, whihc involves multiple trials, the Bernouili disitrbution focuses on just one trial.

Examples:

- Coin Flip (Heads / Tails)
- New Subscriber from watching a Youtube Video (Subscribe / Not Subscribe)
- Pitcher Throwing a Strike (Strike/Ball)

Discrete

Terms to know

- Succeess (p) (also known as probability at one trial)
    - 2.5% of views turn into subscribers
        - P(x=1) = 0.025
        - P(x=0) = 0.0975

```python
import numpy as np
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

np.random.seed(17)
```

## Example #1 Gen Bernoulli from numpy

```python
p_success_subs = 0.025
n_trials_subs = 1
size_videos = 100

numpy_sample = np.random.binomial(n=n_trials_subs,p=p_success_subs, size=size_videos)
numpy_sample
```

## Example #2 Coin Flips in Scipy

```python
p_success_coin = 0.5
size_coin_flips = 100

scipy_samples = bernoulli.rvs(p=p_success_coin, size=size_coin_flips)
scipy_samples
```

## Example #3 Calc PMF 

```python
p_success_coin = 0.5
pmf_success_coin = bernoulli.pmf(k=1, p=p_success_coin)
pmf_failure_coin = bernoulli.pmf(k=0, p=p_success_coin)

print(pmf_success_coin)
print(pmf_failure_coin)
```

## Example #4 CDF

```python
p_success = 0.025
cdf_0 = bernoulli.cdf(0,p_success)
cdf_1 = bernoulli.cdf(1,p_success)

print(cdf_0)
print(cdf_1)
```

```python
plt.figure(figsize=(8,6))
plt.hist(scipy_samples, bins=[-0.5, 0.5, 1.5], color="skyblue", edgecolor="black")
plt.title("Histogram of Cins Filps")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.xticks([0, 1])
plt.show()
```

## PMF Coin Flips

```python
p_success_coin = 0.5
x = np.array([0,1])
pmf = bernoulli.pmf(x, p=p_success_coin)

plt.figure(figsize=(8,6))
plt.bar(x, pmf, width=0.4, color="skyblue", edgecolor="black")
plt.title("PMF Coins Flips")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.xticks([0, 1])
plt.show()
```

## Example 7 CDF of Coin Flip

```python
cdf = bernoulli.cdf(x,p_success_coin)

plt.figure(figsize=(8,6))
plt.step(x, cdf, where='post', label="CDF", color="skyblue", lw=2)
plt.title("CDF Coins Flips")
plt.xlabel("Outcomes")
plt.ylabel("Cumulative Probability")
plt.xticks([0, 1])
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid(True)
plt.show()
```
