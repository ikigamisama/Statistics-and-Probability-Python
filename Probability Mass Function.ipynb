# **Probability Mass Function**

Probability Mass Function - is that gives the probability that a discrete random variable is exactly equal to some value

DISCRETE (if continuous use PDF)

Common Distributions

- Binomial
- Poission
- Geometric
- Uniform

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, geom, randint
```

### Example #1 Binomial

```python
n = 10
p = 0.5
k = np.arange(0, n+1)

pmf = binom.pmf(k, n, p)

plt.figure(figsize=(8,6))
plt.stem(k, pmf)
plt.title("Binomial PMF")
plt.xlabel("Number of Success")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

### Example #2 Poisson

```python
lambda_poisson = 3
k = np.arange(0,15)
pmf = poisson.pmf(k, lambda_poisson)

plt.figure(figsize=(8,6))
plt.stem(k, pmf)
plt.title("Poisson PMF")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

### Example #3 Geometric

```python
p_geom = 0.3
k = np.arange(1,15)
pmf = geom.pmf(k, p_geom)

plt.figure(figsize=(8,6))
plt.stem(k, pmf)
plt.title("Geomteric PMF")
plt.xlabel("Trial Number of First Success")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

### Example #4 Uniform Distribution

```python
k = np.arange(1,11)
pmf = randint.pmf(k, 1, 11)

plt.figure(figsize=(8,6))
plt.stem(k, pmf)
plt.title("Uniform PMF")
plt.xlabel("Trial Number of First Success")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```
