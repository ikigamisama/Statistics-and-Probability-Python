# **Central Limit Theorem**

The central limit theorem states that if we take many samples from any population distribution, the distribution of the sampole means will approach a normal distribution

1. Start with a Population
    a. In this vid we look at populations with Right Skewed Distriubtion, Binomial Distribution, Poisson Distribution

2. Take Samples from the distribution
    a. Ideally More then 25 numbers for a sample
    b. Needs to be at least 30 Samples

3. Find the mean of each Sample

4. Plot the Means on a Histogram

5. More Samples Closer to a Normal Distribution

```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(13)
```

### Example #1 Skwewd Distribution

```python
population_size = 100000
scale = 2
skewed = np.random.exponential(scale=scale, size=population_size)

sns.histplot(skewed, kde=True)
plt.title("Right Skwwed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
skewed_means = []

num_samples = 1000
sample_size_v1 = 30

for i in range(num_samples):
    sample = np.random.choice(skewed, size=sample_size_v1, replace=False)
    sample_mean = sum(sample) / sample_size_v1
    skewed_means.append(sample_mean)

plt.hist(skewed_means, bins=30, edgecolor="black", density=True)
plt.title("Right Skwwed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### Example #2 Binomial Distribution

```python
n_trials = 10
p_success = 0.5
population_size = 10000
binomial_data = np.random.binomial(n_trials, p_success, population_size)

sns.histplot(binomial_data, kde=False)
plt.title("Binomial Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
sample_size_bin = 100
num_samples = 1000
bin_mean = []

for i in range(num_samples):
    sample = np.random.choice(binomial_data, size=sample_size_bin, replace=False)
    sample_mean = sum(sample) / sample_size_bin
    bin_mean.append(sample_mean)

plt.hist(bin_mean, bins=30, edgecolor="black", density=True)
plt.title("Bionimial Means")
plt.xlabel("Bionimial Means")
plt.ylabel("Frequency")
plt.show()

```

### Example #3 Poisson Distribution

```python
lambda_poisson = 3
population_size = 10000
poisson_data = np.random.poisson(lambda_poisson, population_size)
poisson_data

plt.hist(poisson_data, bins=30, edgecolor="black", density=True)
plt.title("Poisson Distribution")
plt.xlabel("Poisson Means")
plt.ylabel("Frequency")
plt.show()
```

```python
sample_size = 30
number_sample = 1000
poission_means = []

for i in range(number_sample):
    sample = np.random.choice(poisson_data, size=sample_size, replace=False)
    sample_mean = sum(sample) / sample_size
    poission_means.append(sample_mean)

plt.hist(poission_means, bins=30, edgecolor="black", density=True)
plt.title("Poission Means")
plt.xlabel("Poission Means")
plt.ylabel("Frequency")
plt.show()

```
