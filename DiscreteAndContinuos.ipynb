# **Discrete VS Continuous Variables**

## **Discrete**

Discrete numbers are countable and distinct. They can take on only specific, separate values (Integer Values)

Example: 

- Number of Runners in a Race: 150 Runners, 20,000 Runers

- Baseball Cards Owner: 500 Cards, 1,100 cards

- Youtube Videos Uploaded: 200 Videos, 400 Videos

## **Continuous**

Continuos numbers can take on any value within a given range. They are measurements (Decimnal Values)

Examples: 

- Miles Ran: 3.1 Miles, 26.2 Miles, 100 Miles

- Weight: 260lbs, 165lbs, 191.2lbs

```python
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

np.random.seed(55)
```

### Example #1 Discrete

```python
dice_rolls = np.random.randint(1,7, size=500)
dice_rolls
```

### Example #2 Continuos

```python
heights = np.random.normal(loc=5.5, scale=0.5, size=500)
heights
```

### Example #3

```python
fig, axes = plt.subplots(1,2, figsize=(14,6))

sns.histplot(dice_rolls, bins=np.arange(0.5, 7.5, 1), ax=axes[0], kde=False)
axes[0].set_title('Discrete Data Dice')
axes[0].set_xlabel('Dice Value')
axes[0].set_ylabel('Frequency')

sns.histplot(heights, ax=axes[1], kde=True, color='green')
axes[1].set_title('Continuos Data Height')
axes[1].set_xlabel('Height')
axes[1].set_ylabel('Density')

plt.tight_layout()
plt.show()
```
