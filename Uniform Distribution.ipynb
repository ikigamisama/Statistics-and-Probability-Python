# **Uniform Distribution**

A uniform distribution is a type of probability distribution in which all outcomes are qually likely within a certain range. Here are some examples

Examples:

- Number on a six sided dice (1/6)

- Random Number generator (Assume 100 random numbers 1/00 probability)


PMF (Probability Mass Function) - Exact Probability that the first success occurs on the k-th trial

CDF (Cumulative distribution function) - Probability that the first success occurs within the first k trials

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import randint

np.random.seed(11)
```

### Example #1 Random Number between 0 and 1

```python
data = np.random.uniform(size=100)
data
```

### Example #2 Dice Rolls


```python
n_rolls = 777
rolls = np.random.randint(1,7, size=n_rolls)

rolls
```

### Example #3 Data points

```python
print(np.mean(rolls))
print(np.std(rolls))
```

### Example #4 Plot histogram

```python
plt.hist(rolls, bins = np.arange(1,8) - 0.5, edgecolor='black', rwidth=0.8)
plt.title('Histogram Dice Rolls')
plt.xlabel("DIce Value")
plt.ylabel("Frequency")
plt.xticks(np.arange(1,7))
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

### Example #5 PMF

```python
dice_distributions = randint(1,7)
value = np.arange(1,7)
pmf_scipy = dice_distributions.pmf(value)

plt.bar(value, pmf_scipy, width=0.5, edgecolor="black", alpha=0.7)
plt.title('PMF Dice Rolls')
plt.xlabel("Dice Value")
plt.ylabel("Probability")
plt.show()
```

### Example #6 CDF

```python
cdf_scipy = dice_distributions.cdf(value)

plt.step(value, cdf_scipy, where='post', label="CDF", color='b', marker='o')
plt.xlabel('Dice Value')
plt.ylabel('Cumulative probability')
plt.title('CDF Dice Rolls')
plt.grid(True)
plt.tight_layout()
plt.show()
```
